```python
import cv2
import pandas as py
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#visualization - first 10 images
num_imgs = 10
rows = 2
cols = 5
plt.figure(figsize = (15, 8))


for i in range(1, num_imgs+1):
    fn = f'{i}.jpeg'
    img = cv2.imread(fn)

    if img is None:
        print(f"error: img {i} not loading\n")
        continue

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(rows, cols, i)
    plt.imshow(img_rgb)

    plt.title(f'Image {i}')    
plt.tight_layout()
plt.show()
```

```python
#Preprocessing 

#Converting to grayscale 
plt.figure(figsize = (15, 8))

for i in range(1, num_imgs+1):
    fn = f'{i}.jpeg'
    img = cv2.imread(fn)

    if img is None:
        print(f"error: img {i} not loading\n")
        continue

    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(rows, cols, i)
    plt.imshow(img_gray)

    plt.title(f'Image {i}')    
plt.tight_layout()
plt.show()
```
